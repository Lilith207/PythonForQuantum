## Hypotheses
The hypothesis is that it will take exponentially long to calculate the results when you expand the number of q-bits to calculated for.  
This is due to the fact that more q-bits will take up more memory and time due to there being more possible states.  
  
Also hypothesized is that increasing theta will affect the time required to calculate a lot harder then increasing the node q-bits.  
This is due to the fact that the one step circuit will be used a lot more when changing theta. See below for how much, n is adjustable to the number of theta q-bits.  

The last hypothesis is that increasing the iteration will increase the time taken in a linear manner.

```python
n = 5
GateTotal = 0
for i in range(n):
    GateTotal += 2**i

print(GateTotal)
```

## Test 3: increasing iteration count
So here iterations will be increased from 2 to 10

Iterations:   2, 3, 4, 5, 6, 7, 8, 9, 10.  
Times measured: 15.2, 15.8, 16.7, 17.3, 18.5, 19.0, 20.7, 20.5, 22.7.

```python
from matplotlib import pyplot as py

Nums = [2, 3, 4, 5, 6, 7, 8, 9, 10]
NodeNumT =   [13.4, 13.4, 15.5, 17.1, 18.9, 20.7, 19.4, 21.0, 21.9]
ThetaQbitT = [14.1, 14.0, 14.9, 16.6, 23.1, 33.7, 41.9, 78.8, 142.9]
IterT =      [15.2, 15.8, 16.7, 17.3, 18.5, 19.0, 20.7, 20.5, 22.7]
py.plot(Nums, NodeNumT)
py.plot(Nums, ThetaQbitT)
py.plot(Nums, IterT)
py.show()
```

## Conclusion
As we can see in the above graph, both node number and iteration will increase the time taken to calculate linearly. But increasing the theta qbits increases the time exponentially.
